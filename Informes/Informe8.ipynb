# ***EJERCICIO DIJKSTRA***

***1. OBJETIVOS*** 

* Aplicar el algoritmo de Dijkstra para determinar la ruta de menor costo desde un nodo inicial (Start) hasta un nodo destino (Finish) en un grafo ponderado dirigido.


***2. INTRODUCCIÓN***

El algoritmo de Dijkstra es uno de los métodos más importantes y ampliamente utilizados en ciencias de la computación para resolver el problema del camino más corto en grafos ponderados. Tiene multiples aplicaciones desde sistemas de navegación GPS y enrutamiento de redes de comunicación, hasta la planificación de rutas logísticas y la optimización de procesos industriales.

El ejercicio tiene como propósito implementar el algoritmo de Dijkstra de manera manual sobre un grafo dirigido con pesos positivos, siguiendo una metodología estructurada que permita visualizar y comprender cada paso del proceso de búsqueda. A través de la construcción de una tabla de ejecución detallada, se documenta cómo el algoritmo selecciona iterativamente el nodo no procesado con menor costo acumulado, explora sus vecinos, actualiza las distancias cuando encuentra caminos más cortos, y mantiene un registro de los nodos padres para reconstruir la ruta óptima.
